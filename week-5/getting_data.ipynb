```python
import textract
import math
import pandas as pd

dec_map = {1:22, 2:18, 3:18,4:21, 5:22, 6:7, 7:41, 8:26, 9:22, 10:18, 11:17, 
           12:10, 13:14,14:10, 15:14, 16:13, 17:20, 18:27, 19:29, 20:25, 
           21:24, 22:26, 23:23, 24:19, 25:19}
```

```python
decs = []
for year in dec_map:
    for i in range(1, dec_map[year]+1):
        fpath = 'decisions/cp'+str(year) +'/'+ str(year)+'-'+ str(i)+'.pdf'
        decs.append(fpath)
```

```python
decs
```

```python
def clean_text(file_name):
    '''
    Open a pdf file, extract the text and return it.
    
    Input:
       file_name, string
    Output:
       string of text
          
    '''
    text = textract.process(file_name).decode('utf-8')
    text = text.replace("\n", "")
    text = text.replace("■", "")
    return text
```

```python
text_by_dec = pd.DataFrame()
texts = []

for dec in decs:
    
    text = clean_text(dec)
    
    texts.append(text)

    print(dec,'done')
```

```python
text_by_dec = pd.DataFrame()
text_by_dec['Decision'] = decs
text_by_dec['Texts'] = texts
```

```python
text_by_dec['Texts'][0]
```

```python
import pickle


with open('text_by_dec.pickle', 'wb') as handle:
    pickle.dump(text_by_dec, handle, protocol=pickle.HIGHEST_PROTOCOL)

with open('text_by_dec.pickle', 'rb') as handle:
    b = pickle.load(handle)
```

```python
text_by_dec == b
```

```python
import lucem_illud_2020
```

```python
text_by_dec['tokenized_sents'] = text_by_dec['Texts'].apply(lambda x: [lucem_illud_2020.word_tokenize(s) for s in lucem_illud_2020.sent_tokenize(x)])
text_by_dec['normalized_sents'] = text_by_dec['tokenized_sents'].apply(lambda x: [lucem_illud_2020.normalizeTokens(s) for s in x])
```
