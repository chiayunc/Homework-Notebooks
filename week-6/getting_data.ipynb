```python
def get_page_text(pdf_document, topic):
    #pdf_document = 'climate_labeled_data/adaptation.pdf'
    pdf = PdfFileReader(pdf_document)
    l_text = []
    l_id = []
    l_cat = []


    for page in range(pdf.getNumPages()):

        current_page = pdf.getPage(page)
        text = current_page.extractText()
        text = text.replace("\n", "")
        l_text.append(text)
        l_id.append(topic +'-'+ str(page+1))
        l_cat.append(topic)
        
    return  l_id,l_cat ,l_text
        
        
```

```python
all_id = []
all_cat = []
all_text = []


for i,f in enumerate(files):
    l_id,l_cat ,l_text = get_page_text(f, t[i])
    all_id += l_id
    all_cat += l_cat
    all_text += l_text
    
    


    
    
```
